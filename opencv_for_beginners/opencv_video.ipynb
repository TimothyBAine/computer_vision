## Working with Video Captures

```python
# Establish capture
cap = cv2.VideoCapture('video1.webm')

# Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = height = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FRAME_COUNT)

# Video Writer
video_writer = cv2.VideoWriter(os.path.join("output.mp4"), cv2.VideoWriter_fourcc('P', 'I', 'M', '1'), 
                              fps, (width, height), isColor=False) #isColor = False because we a saving a grayscale video

# Loop through each frame
for frame_idx in range (int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    # Read the frame
    ret, frame = cap.read()
    
    # Transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    # Show image
    cv2.imshow('Video Player', gray)
    
    # Write out the frame
    video_writer.write(gray)
    
    # Render
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
#Close everything
cap.release()
cv2.destroyAllWindows()
```
