### Importing Dependencies

```python
import cv2
import os
```

### Reading in images

```python
bike_path = os.path.join('data', 'bike.jpg')
cola_path = os.path.join('data', 'cola.jpg')
fruit_path = os.path.join('data', 'fruit.jpg')
```

```python
bike = cv2.imread(bike_path)
cola = cv2.imread(cola_path)
fruit = cv2.imread(fruit_path)

print(f'Bike shape: {bike.shape}\nCola Shape: {cola.shape}\nFruit shape: {fruit.shape}')
```

### Rendering images

```python
import matplotlib.pyplot as plt
```

```python
# show image
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.suptitle('Rendered Images')
ax1.imshow(bike)
ax2.imshow(cola)
ax3.imshow(fruit)
```

The colors seem to be different when shown with matplotlib.
This is because of the fact that where as cv2 reads in the images in a BGR format, matplotlib reads it as an RGB.
The colors are therefore reversed

```python
#Recoloring images
bike_recolor = cv2.cvtColor(bike, cv2.COLOR_BGR2RGB)
cola_recolor = cv2.cvtColor(cola, cv2.COLOR_BGR2RGB)
fruit_recolor = cv2.cvtColor(fruit, cv2.COLOR_BGR2RGB)
```

```python
# show image
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.suptitle('Rendered Images')
ax1.imshow(bike_recolor)
ax2.imshow(cola_recolor)
ax3.imshow(fruit_recolor)
```

### Writing images


```python
#Gray scaling images
bike_gray = cv2.cvtColor(bike, cv2.COLOR_BGR2GRAY)
cola_gray = cv2.cvtColor(cola, cv2.COLOR_BGR2GRAY)
fruit_gray = cv2.cvtColor(fruit, cv2.COLOR_BGR2GRAY)

cv2.imwrite('bike_gray.jpg', bike_gray)
cv2.imwrite('cola_gray.jpg', cola_gray)
cv2.imwrite('fruit_gray.jpg', fruit_gray)
```
